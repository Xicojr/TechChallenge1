```python
import pandas as pd
import seaborn as sns
```

```python
dados_vinhos = pd.read_csv("https://github.com/Xicojr/TechChallenge1/raw/main/Dados/wine-consumption-per-person%20(1).csv", sep=",")
dados_vinhos.head()
```

```python
dados_vinhos.columns
```

```python
dados_vinhos = dados_vinhos.rename(columns= {'Indicator:Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol) - Beverage Types:Wine':'media_consumo'})
```

```python
dados_vinhos.columns
```

```python
dados_vinhos
```

```python
#Filtrar por valores específicos
dados_vinhos2 = dados_vinhos[dados_vinhos["Year"]== 2008]
dados_vinhos2 = dados_vinhos2.append(dados_vinhos[dados_vinhos["Year"]== 2019])
```

```python
filtered_df = dados_vinhos2[dados_vinhos2['Year'].isin([2008, 2019])]

pivot_df = filtered_df.pivot(index='Entity', columns='Year', values='media_consumo').reset_index()

pivot_df
```

```python
#pivot_df['Absolute_change'] = pivot_df.pct_change(axis=1)[2019]

pivot_df['Absolute_change'] = pivot_df[2019]-pivot_df[2008]

pivot_df
```
